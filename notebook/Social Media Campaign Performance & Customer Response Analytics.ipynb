```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df1 = pd.read_csv("C:/Users/roshn/Downloads/archive (1)/Facebook Metrics of Cosmetic Brand.csv")
```

```python
df1.head()
```

```python
cols_to_keep = [
    'Type',
    'Paid',
    'Post Month',
    'Post Weekday',
    'Post Hour',
    'Lifetime Post Total Reach',
    'Lifetime Post Total Impressions',
    'like',
    'comment',
    'share',
    'Total Interactions'
]

df = df1[cols_to_keep]
```

```python
df.head()
```

```python
df.info()
```

```python
# Fill NaNs in engagement columns and convert to int
df['like'] = df['like'].fillna(0).astype(int)
df['share'] = df['share'].fillna(0).astype(int)

# Paid: fill missing with 0 (assume organic) and convert to int
df['Paid'] = df['Paid'].fillna(0).astype(int)
```

```python
df.info()
##df.dtypes
```

```python
df.head()
```

```python
df['Total_Engagement'] = df['like'] + df['comment'] + df['share']
```

```python
df[['like', 'comment', 'share', 'Total_Engagement']].head()
```

```python
df['Engagement_Rate'] = df['Total_Engagement'] / df['Lifetime Post Total Reach']
```

```python
df.head()
```

```python
df['Engagement_Rate_Pct'] = df['Engagement_Rate'] * 100
```

```python
(df['Total_Engagement'] == df['Total Interactions']).value_counts()
```

```python
df['Engagement_Rate_Pct'] = (df['Engagement_Rate'] * 100).round(2)
```

```python
df.head()
```

```python
#Create readable labels for Paid Vs Organic
df['Paid_Flag'] = df['Paid'].map({0: 'Organic', 1: 'Paid'})
```

```python
df['Paid_Flag'].value_counts()
```

```python
paid_summary = (
    df.groupby('Paid_Flag')
      .agg(
          Avg_Reach=('Lifetime Post Total Reach', 'mean'),
          Avg_Impressions=('Lifetime Post Total Impressions', 'mean'),
          Avg_Total_Engagement=('Total_Engagement', 'mean'),
          Avg_Engagement_Rate_Pct=('Engagement_Rate_Pct', 'mean'),
          Post_Count=('Paid_Flag', 'count')
      )
      .round(2)
)

paid_summary
## we see organic posts have lower reach by higher engagement efficiency, 
## paid posts have higher reach lower engagement efficiency
## Paid posts increased reach significantly, but organic posts achieved higher engagement rates, 
## indicating stronger contentâ€“audience fit despite lower exposure. "Paid improves distribution; organic drives efficiency"
```

```python
paid_summary[['Avg_Reach', 'Avg_Engagement_Rate_Pct']]
```

```python
paid_summary.dtypes
```

```python
Type_summary = (
    df.groupby('Type')
      .agg(
          Avg_Reach=('Lifetime Post Total Reach', 'mean'),
          Avg_Impressions=('Lifetime Post Total Impressions', 'mean'),
          Avg_Total_Engagement=('Total_Engagement', 'mean'),
          Avg_Engagement_Rate_Pct=('Engagement_Rate_Pct', 'mean'),
          Post_Count=('Paid_Flag', 'count')
      )
      .round(2)
)
Type_summary
```

```python
Type_summary[["Avg_Total_Engagement","Avg_Engagement_Rate_Pct","Post_Count"]]
##engagement rate >> This shows efficiency (who performs best per unit of reach)
##total engagement >> This shows volume (who gets more reactions overall)
```

```python
##Video posts achieve higher reach, while photo posts generate more engagement per user, 
##suggesting photos drive stronger audience interaction despite lower exposure.
```

```python
paid_type_summary = (
    df.groupby(['Paid_Flag', 'Type'])
      .agg(
          Avg_Total_Engagement=('Total_Engagement', 'mean'),
          Avg_Engagement_Rate_Pct=('Engagement_Rate_Pct', 'mean'),
          Post_Count=('Type', 'count'),
          Reach_Count=('Lifetime Post Total Reach','mean')
      )
      .round(2)
      .reset_index()
)

paid_type_summary
```
